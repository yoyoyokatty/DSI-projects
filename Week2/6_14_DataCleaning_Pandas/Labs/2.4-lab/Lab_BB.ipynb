```python
def changeRandB(value):
    if value == "R & B":
        return "R&B"
    else:
        return value

def changerr(value):
    if value == "Rock'n'roll":
        return "Rock"
    else:
        return value

bb["genre"] = bb["genre"].apply(changerr)
bb["genre"] = bb["genre"].apply(changeRandB)
```

```python
bb['genre'].value_counts()
```

```python
bb["date.entered"] = pd.to_datetime(bb["date.entered"])
```

```python
bb["date.peaked"] = pd.to_datetime(bb["date.peaked"])
```

```python
bb
```

```python
def AMremover(string):
    if " AM" in string:
        newstring = string.replace(" AM", "")
        if "," in newstring:
            newnewstring = newstring.replace(",", ":")
        return newnewstring

bb["time"] = bb["time"].apply(AMremover)
```

```python
bb["time"] = pd.to_timedelta(bb["time"])
```
