```python
#reformat urls so that they can be read into wikipedia api
list_wiki = []

for url in list_artists:
    url = url.replace("https://en.wikipedia.org/wiki/", "").replace("_", " ").replace("https://en.wikipedia.org/w/index.php?title=", "")
    list_wiki.append(url)
```

```python
#pull all links in each url page into a list - the only links on these List type pages are artist names so that will do.

#some of the links don't work, so these exceptions are printed and treated separately
content_list = []

for i in list_wiki:
    try:
        genre_artist = wikipedia.page(i)
        article = genre_artist.links
        content_list.append(article)
    except:
        print i 
        pass
```

```python
#left over list of pages that don't work - continue to append to content_list

new_list_wiki = ['List of punk rock bands, 0\u2013K', 'List of punk rock bands, L\u2013Z', 'List of hard rock musicians (A\u2013M)', 'List of hard rock musicians (N\u2013Z)', 'List of R&B musicians', 'List of disco artists','List of disco artists (L\u2013R)','List of Euro disco artists','List of disco artists (F\u2013K)','List of disco artists (S\u2013Z)','List of disco artists (A\u2013E)']

for i in new_list_wiki:
    try:
        new_genre_artist = wikipedia.page(i)
        new_article = new_genre_artist.links
        content_list.append(new_article)
    except:
        print i 
        pass
```

```python
#iterate one more time to append onto content_list

final_list_wiki = ["List of punk rock bands, 0-K", "List of punk rock bands, L-Z", "List of hard rock musicians (A-M)", "List of hard rock musicians (N-Z)","List of disco artists (L-R)", "List of disco artists (F-K)","List of disco artists (S-Z)","List of disco artists (A-E)"]

for i in final_list_wiki:
    try:
        final_genre_artist = wikipedia.page(i)
        final_article = final_genre_artist.links
        content_list.append(final_article)
    except:
        print i 
        pass
```

```python
#flatten the content_list because right now its a list of a list

flattened = [val for sublist in content_list for val in sublist]
```

```python
#reformat artist names

flattened_new = [i.encode("ascii", "ignore") for i in flattened]
```

```python
#continue to reformat - reformat artist names so its names only

import re

finallist=[re.sub("\s\([^)]*\)","", name) for name in flattened_new]
```

```python
#reformat list so that full names are appended with "+" in between to feed into iTunes API

#note that some names on this include names that are not artists because this is all links on a page so publications /
# etc could be included

finalfinalist = []

for term in finallist:
    if " " in term:
        finalfinalist.append(term.replace(" ", "+"))
    else:
        finalfinalist.append(term)
```

### Begin to append artist list from wikipedia to iTunes

```python
url_template = "https://itunes.apple.com/search?term={}&entity=album"

url_list = []

for item in finalfinalist:
    newurl = url_template.replace("{}", item)
    url_list.append(newurl)
```

```python
import pickle

with open('url_list.pkl', 'w') as picklefile:
    pickle.dump(url_list, picklefile)
```

### Take full iTunes Pull into Data Table

```python
#open each json file and put it into a data table

#note that some errors will occur because some json files are empty given that the final content list was used with /
#all links on list pages
import json
import pandas as pd

empty_list = []
for i in range(1,29216):
    try:
        with open("../json_data_dump/"+str(i)+'.txt') as data_file:    
            i = json.load(data_file)
            empty_list.append(i)
    except:
        print i
        pass
    
appended_list = []

for item in empty_list:
    for i in item["results"]:
        i = pd.Series(i)
        appended_list.append(i)
    testtest = pd.concat(appended_list, axis = 1)
    testtest = testtest.T
```

```python
#begin to format
testtest.applymap(lambda x: x.encode("ascii", "ignore") if type(x) == "str" else x)
```

```python
#finalize dataframe
df = testtest[["artistName", "artworkUrl100", "releaseDate", "primaryGenreName", "collectionName", "country"]]
```

```python
#adjust decades
from datetime import datetime

df["year"] = df["releaseDate"].apply(lambda x: datetime.strptime(x[0:4], "%Y").year)
```

```python
import numpy as np

def decader(boops):
    if boops in range(1970,1980):
        return int(1970)
    elif boops in range(1980, 1990):
        return int(1980)
    elif boops in range(1990, 2000):
        return int(1990)
    elif boops in range(2000, 2010):
        return int(2000)
    elif boops in range(2000, 2016):
        return int(2010)
    else:
        return np.nan
```

```python
df["decade"] = df["year"].apply(decader)
```
